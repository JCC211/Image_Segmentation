```python
import skimage as sk
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load image and display
car = "car.jpg"
car = sk.io.imread(car)
img = car
plt.title('Car')
plt.imshow(car), car.shape
```

```python
# image processing
car = car / 255
car = car.reshape(-1,3)
car.shape
```

```python
# k-means with different n_clusters
km = KMeans(n_clusters=2,random_state=33).fit(car)
km2 = KMeans(n_clusters=3,random_state=33).fit(car)
km3 = KMeans(n_clusters=4,random_state=33).fit(car)
km4 = KMeans(n_clusters=5,random_state=33).fit(car)
centers = km.cluster_centers_
labels = km.labels_
centers2 = km2.cluster_centers_
labels2 = km2.labels_
centers3 = km3.cluster_centers_
labels3 = km3.labels_
centers4 = km4.cluster_centers_
labels4 = km4.labels_
print(f"Cluster centers: {km.cluster_centers_}\nData labels: {km.labels_.shape}\n{km.labels_}")
```

```python
# reshape segmented images
km_img = centers[labels.flatten()]
km_img = km_img.reshape((img.shape))
km_img2 = centers2[labels2.flatten()]
km_img2 = km_img2.reshape((img.shape))
km_img3 = centers3[labels3.flatten()]
km_img3 = km_img3.reshape((img.shape))
km_img4 = centers4[labels4.flatten()]
km_img4 = km_img4.reshape((img.shape))
km_img.shape
```

```python
# plot results
plt.subplot(2,2,1)
plt.imshow(km_img)
plt.subplot(2,2,2)
plt.imshow(km_img2)
plt.subplot(2,2,3)
plt.imshow(km_img3)
plt.subplot(2,2,4)
plt.imshow(km_img4)
```
